```python
from sklearn.preprocessing import OrdinalEncoder
from sklearn.impute import SimpleImputer

str_cols = ["وضعیت دانشجو", "نوع دوره پزشکی", "رشته تحصیلی"]  # String/categorical columns
encoder = OrdinalEncoder()
data[str_cols] = encoder.fit_transform(data[str_cols])

imputer = SimpleImputer(strategy="most_frequent")
data[str_cols] = pd.DataFrame(
    imputer.fit_transform(data[str_cols]),
    columns=data[str_cols].columns,
    index=data[str_cols].index
)
```

```python
import pandas as pd

# Retrieve the categories from the fitted encoder
mapping_dict = {}
for i, col in enumerate(str_cols):
    # The encoder stores categories in a list of arrays
    categories = encoder.categories_[i]
    # Create a dictionary where the index is the integer and the value is the original name
    mapping_dict[col] = {idx: label for idx, label in enumerate(categories)}

# Display the mappings clearly
for column, mapping in mapping_dict.items():
    print(f"--- Mapping for '{column}' ---")
    for integer_val, original_val in mapping.items():
        print(f"  {integer_val} -> {original_val}")
    print()
```

```python
import lightgbm as lgb
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd
import json
import onnx
import onnxmltools
from onnxmltools.convert.common.data_types import FloatTensorType
import matplotlib.pyplot as plt

# Optimized model for small dataset and wider prediction range
model = lgb.LGBMRegressor(
    n_estimators=500,
    learning_rate=0.1,
    num_leaves=15,
    max_depth=5,
    min_child_samples=5,  # Important for small datasets
    boosting_type='gbdt',
    objective='regression',
    random_state=42,
    n_jobs=-1,
    importance_type='gain'
)

def run_model(x_data, y_data, num):
    x_train, x_val, y_train, y_val = train_test_split(x_data, y_data, test_size=0.2, random_state=42)

    scaler = StandardScaler()
    x_train = pd.DataFrame(scaler.fit_transform(x_train), columns=x_train.columns, index=x_train.index)
    x_val = pd.DataFrame(scaler.transform(x_val), columns=x_val.columns, index=x_val.index)

    model.fit(
        x_train, y_train.values.ravel(),
        eval_set=[(x_val, y_val.values.ravel())],
        eval_metric='rmse',
        callbacks=[lgb.early_stopping(stopping_rounds=30), lgb.log_evaluation(period=0)]
    )

    # Check prediction range
    preds = model.predict(x_val)
    print(f"Model {num} - Actual Range: [{y_val.values.min():.2f}, {y_val.values.max():.2f}] - Predicted Range: [{preds.min():.2f}, {preds.max():.2f}]")

    scaling_params = {
        "mean": scaler.mean_.tolist(),
        "std": scaler.scale_.tolist(),
        "feature_names": x_train.columns.tolist()
    }
    with open(f"scaling_params{num}.json", "w") as f: json.dump(scaling_params, f)

    n_features = x_train.shape[1]
    initial_type = [('float_input', FloatTensorType([None, n_features]))]
    onnx_model = onnxmltools.convert_lightgbm(model, initial_types=initial_type, target_opset=12)
    onnx.save_model(onnx_model, f"LGBMModel{num}.onnx")

    return model, scaler

# Re-train models
for label, x_df, y_df in [
    (41, x41_data, y41_data), (32, x32_data, y32_data),
    (31, x31_data, y31_data), (22, x22_data, y22_data),
    (21, x21_data, y21_data), (12, x12_data, y12_data)
]:
    run_model(x_df, y_df, label)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

def get_predictions_table(x_data, y_data, model_name):
    # Replicate the split used in training
    x_train, x_val, y_train, y_val = train_test_split(x_data, y_data, test_size=0.2, random_state=42)

    # Scale x_val using a local scaler (matching training logic)
    scaler = StandardScaler()
    scaler.fit(x_train)
    x_val_scaled = pd.DataFrame(scaler.transform(x_val), columns=x_val.columns, index=x_val.index)

    # Generate predictions
    preds = model.predict(x_val_scaled)

    # Create comparison table starting with features (x_val)
    results = x_val.copy().reset_index(drop=True)

    # Add target and predictions
    results['Actual_Score'] = y_val.values.flatten()
    results['Predicted_Score'] = preds
    results['Difference'] = results['Actual_Score'] - results['Predicted_Score']

    return results

# Dictionary to store results
all_results = {}

# Iterate through each dataset to show comparisons
for label, x_df, y_df in [
    (41, x41_data, y41_data), (32, x32_data, y32_data),
    (31, x31_data, y31_data), (22, x22_data, y22_data),
    (21, x21_data, y21_data), (12, x12_data, y12_data)
]:
    # Ensure the 'model' variable is correct for this label
    current_model, _ = run_model(x_df, y_df, label)
    table = get_predictions_table(x_df, y_df, label)
    print(f"\n--- Comparison Table for Model {label} (First 10 rows) ---")
    display(table.head(10))
    all_results[label] = table
```
